```python
# Question 4

# I have a python script that I use to calculate student grade classifications.
# It takes an input of their results, calculates their average, and then outputs
# their grade classification. ​

# The classifications are as follows:​
#    80+ : A​
#    70-80 : B​
#    60-70 : C​
#    50-60 : D​
#    0-49 : Fail

# There are a few potential problems here involving edge cases.

def get_classification(average):      # For purposes of clarity, the students average is calculated outside the function.

  if average >= 80:
    return "A"
  elif average >= 70:
    return "B"
  elif average >= 60:
    return "C"
  elif average >= 50:
    return "D"

# If instead of using >= signs, we had instead just used > signs, there would be
# issues with classifying results on the edge. Some assertion tests to check
# whether our edge cases are working as intended would look like the following.

assert get_classification(100) == 'A'
assert get_classification(80) == 'A'
assert get_classification(79) == 'B'
assert get_classification(70) == 'B'
assert get_classification(69) == 'C'
assert get_classification(60) == 'C'
assert get_classification(50) == 'D'
assert get_classification(49) == 'Fail'
```
